7. El siguiente script multiplica la variable p por 0.1 hasta que su valor sea menor que la tolerancia tol o hasta que alcance el número máximo de iteraciones permitidas

```
# Break
# Definición de variables
MaxIter  = 15     # número máximo de iteraciones
contador = 0      # contador
Tol      = 1e-7   # tolerancia
p = 1.0           # vaalor inicial de la variable
# Ciclo
for i in range(0,MaxIter-1):
    contador +=  1
    p = p*0.1
    print('contador = {}, p = {} '.format(contador,p)) 
    if p < Tol:
        break
```

Este código usa un ciclo for para el proceso iterativo y para con un if. Sustituye el if y el for por un ciclo while, hasta que p sea menor que la tolerancia dada y que no pase el número máximo de iteraciones
