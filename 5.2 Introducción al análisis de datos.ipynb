# Visualización con pandas

### DataFrame.plot
Los Data Frames se pueden graficar directamente con funciones pre-definidas en pandas.

A continuación, revisaremos algunas opciones.

```python
# DataFrame
import pandas as pd
import numpy as np
ds = pd.DataFrame({'A':np.random.randn(10000), 'B':np.random.randn(10000), 'C':np.random.randn(10000), 'D':np.random.randn(10000)*3})
ds
```

```python
ds_cumsum = ds.cumsum() # Suma culumativa de los df
ds_cumsum
```

```python
ds_cumsum['index'] = list(range(len(ds_cumsum)))
ds_cumsum['index'] 
```

```python
small_ds = ds.sample(10) # muestra de los resultados
small_ds
```

## Graficas básicas
La función básica para graficar un dataframe es

DataFrame.plot(), que grafíca lineas

```python
ds
```

```python
ds_cumsum.plot()
```

```python
columns = ["A","B","C","D"]
ds_cumsum[columns].plot()
```

```python
columns = ["A","D"]
ds_cumsum[columns].plot()
```

```python
ds_cumsum.drop('index', axis=1).plot()
```

Los tipos de graficas osn los sigientes:

`kind` :
- `'line'` : line plot (default)
- `'bar'` : vertical bar plot
- `'barh'` : horizontal bar plot
- `'hist'` : histogram
- `'box'` : boxplot
- `'kde'` : Kernel Density Estimation plot
- `'density'` : same as 'kde'
- `'area'` : area plot
- `'pie'` : pie plot
- `'scatter'` : scatter plot
- `'hexbin'` : hexbin plot

```python
ds_cumsum[0:10].plot('index','B', kind = 'scatter')
```

```python
ds_cumsum.plot('A','B', kind = 'kde')
```

```python
ds_cumsum.plot(x='index',y='A', color='magenta')
```

```python
ds_cumsum.iloc[1:10].plot(x='index',y='A', marker="d",markersize=5,color='magenta')
```

```python
ds_cumsum.drop('index', axis=1).plot(subplots=True, figsize=(10,10))
```

```python
ds_cumsum.A.plot()
ds_cumsum.B.plot()
```

```python
ds_cumsum.A.plot()
ds_cumsum.B.plot(secondary_y=True)
```

# Histogramas

```python
ds.A.plot.hist()
```

```python
# Varios histogramas
ds_hist = pd.DataFrame({'A':np.random.randn(1000)-1,'B':np.random.randn(1000),'C':np.random.randn(1000)+1}) 
ds_hist.plot.hist()
```

```python
# El parametro alfa nos ayuda a cambiar la transparencia de las graficas
ds_hist.plot.hist(alpha=.4)
```

```python
# También se pueden encimar los histogramas
ds_hist.plot.hist(stacked=True, colormap='cool')
```

```python
# hacer subplpots
ds_hist.plot.hist(subplots=True, layout=(3,1), figsize=(16,5), colormap='cool')
```

## Diagrama hexadecimal (hexbin)
Un diagrama de hexadecimal es útil para representar la relación de 2 variables numéricas cuando tiene muchos puntos de datos. Sin la superposición de los puntos, la ventana de trazado se divide en varios cubos hexagonales. El color de cada cubo hexagonal indica el número de puntos que contiene. 
Algunos mapas de color se puenen encontrar en el siguente enlace: [mapas](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
ds.plot.hexbin(x='A',y='B', gridsize=25, colormap='cividis')
```

```python
ds.plot('A','B', kind = 'scatter');
```
